```python
# Creating hardcoded dictionary for location and file path of each flower type.
flowers_images_dict = {
    'roses': list(data_dir.glob('roses/*.jpg')),
    'daisy':list(data_dir.glob('daisy/*.jpg')),
    'dandelion':list(data_dir.glob('dandelion/*.jpg')),
    'sunflowers':list(data_dir.glob('sunflowers/*.jpg')),
    'tulips':list(data_dir.glob('tulips/*.jpg')),
              }
            
```

```python
# Creating hardcoded dictionary for loaction and file path of each flower type.
flowers_labels_dict = {
    'roses': 0,
    'daisy':1,
    'dandelion':2,
    'sunflowers':3,
    'tulips':4,
              }
```

```python
# Using for loops to generate X and y arrays.
X,y = [],[]
for flower_name, image_paths in flowers_images_dict.items(): # returns (key,values)--> (flower_name, image_paths)
    for i in image_paths:
        img = cv2.imread(str(i))
        resized_image = cv2.resize(img,(IMAGE_SHAPE))
        X.append(resized_image)
        y.append(flowers_labels_dict[flower_name])
```

```python
X[0]
```

```python
X= np.array(X) # X was a list above. Converting to numpy
y= np.array(y) # y was num
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0) # Default ratio at 0.25 
```

```python
X_train_scaled = X_train/255
X_test_scaled = X_test/255
```

**We see that the classes we require are not present in the image_net_labels.txt. So even if we train the model based on the pretrained data, we cannot classify some of the classes.**

**The only solution to this problem is:**
        
        1. To train our own model from scratch - Too expensive
        2. To use the pretrained model excluding the last layer. We will extract the features only.
           Then Our classes can be trained- Less computation and cheap.

```python
len(flowers_labels_dict) # Value to be inserted in next cells model
```

```python
# Follow the above cell for reference. Model is built in next cell.Simply preparing the inputs here.
feature_extractor_model = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
# 'classification' replaced with 'feature_vector'
pretrained_model_without_top_layer = hub.KerasLayer(feature_extractor_model, input_shape=(224,224,3),trainable = False)
#trainable = False means do not train. we will train the model based on our classes.
```

```python
model = keras.Sequential([
    pretrained_model_without_top_layer, # from above cell
    keras.layers.Dense(len(flowers_labels_dict)) # Linear activation. No softmax is used.
])
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=keras.losses.SparseCategoricalCrossentropy(from_logits = True), ## True logits- (sigmoid/softmax not used - Linear output desired in form of numbers - no probabilities)        
              metrics=['accuracy'])
```

```python
# Initialy we trained the model on 30 epochs P11- Accuracy reached for training set = 0.9517
model.fit(X_train_scaled,y_train,epochs = 5) 
```

```python
from keras.models import load_model
# from tensorflow import keras
# model.save('model.h5')
model_new = load_model('model.h5')
```

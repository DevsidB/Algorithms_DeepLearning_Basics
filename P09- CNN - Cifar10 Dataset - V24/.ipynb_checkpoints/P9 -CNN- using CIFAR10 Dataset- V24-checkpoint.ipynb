## Predicting using CNN'S

    CNN's are better at classification than ANN as they can detect the features from any location throughout the images.
    Also Convolution and pooling reduces the computing time by reducing the pixel size and also not loosing the required info.
    This reduces the computing time for our model to predict the results.

```python
0.19999999999999998 + 0.3333333333333333 + 0.4666666666666666
```

```python
# %%timeit -n1 -r1
cnn = models.Sequential ([
    
    layers.Conv2D(filters = 32, kernel_size = (3,3),activation = 'relu',input_shape = (32,32,3)), 
#     filters denote number of different patterns to detect. 
#     It figures out on its own just like a loopy circle in number 9, straight line or a diagonal line, curves edges etc.
#     kernel_size denotes the grid size for creating partterns. Can be any number 3x3 4x4 etc
    layers.MaxPooling2D((2,2)),  
    # Max pooling and average pooling are the two popular methods. 
    # In max pooling of (2,2) from a grid of 2x2 a max and adds to a mapping.
    
    layers.Conv2D(filters = 64, kernel_size = (3,3),activation = 'relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64,activation ='relu'),
    layers.Dense(10,activation = 'softmax')
    ])

cnn.compile (optimizer= 'adam',
          loss = 'sparse_categorical_crossentropy',
          metrics = ['accuracy'])

cnn.fit(X_train,y_train,epochs=10)
```

```python
from keras.models import load_model
# from tensorflow import keras

# Model can be saved without using the fit function but will build a dumb model. In this case accuracy: 0.0890(epochs reduced to 1)

# cnn.save('cnn_model.h5')
# cnn_model = keras.models.load_model('cnn_model.h5')
cnn_model = load_model('cnn_model.h5')
```

```python
# We will be using these functions inside tf pipelines. So make sure they are compatible. Else error will show up.
#.split needs to be called in tf.strings.split (only .split is not there)
def get_label(file_path):
    import os
    return tf.strings.split(file_path,os.path.sep)[-2]  
```

```python
def get_path(file_path):
    import os
    return tf.strings.split(file_path,os.path.sep)[-1]  
```

```python
get_label
```

```python
get_label('images\\dog\\43 Best Small Dog Breeds \xe2\x80\x94 Popular Toy (1).jpg')
```

```python
# Generating Combined image and labels
def process_image(file_path):
    label = get_label(file_path)
    path_name = get_path(file_path)
    img = tf.io.read_file(file_path)
    img = tf.image.decode_jpeg(img)
    img = tf.image.resize(img,[128,128])
    return img,label,path_name
```

```python
for image,label,path_name in images_ds.map(process_image):
    print(image)
    print(label)
    print(path_name)
```
